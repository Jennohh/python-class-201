```python
john_adams_df = df[df["School Name"] == "John Adams High School"]
```

```python
#dataframe for queens schools
queens_df = df[df["School DBN"].str.contains("Q")]
```

```python
#average percentage for each comparison
#cleaning
```

```python
df["Percent Scoring 80 or Above"] = pd.to_numeric(df["Percent Scoring 80 or Above"], errors="coerce")
```

```python
df_clean = df.dropna(subset=["Percent Scoring 80 or Above"])
```

```python
john_adams_df["Percent Scoring 80 or Above"] = pd.to_numeric(john_adams_df["Percent Scoring 80 or Above"], errors="coerce")
john_adams_80 = john_adams_df["Percent Scoring 80 or Above"].mean()
```

```python
john_adams_80 = john_adams_df["Percent Scoring 80 or Above"].mean()
```

```python
queens_df["Percent Scoring 80 or Above"] = pd.to_numeric(queens_df["Percent Scoring 80 or Above"], errors="coerce")
queens_80 = queens_df["Percent Scoring 80 or Above"].mean()
```

```python
nyc_80 = df_clean["Percent Scoring 80 or Above"].mean()
```

```python
boroughs_80 = df["Percent Scoring 80 or Above"].astype(float).mean()
```

```python
john_adams_80, queens_80, nyc_80
```

```python
categories = ["John Adams", "Queens", "Five Boroughs"]
values = [john_adams_80, queens_80, boroughs_80]
```

```python
#sorting in descending order
sorted_indices = np.argsort(values)[::-1]
sorted_categories = [categories[i] for i in sorted_indices]
sorted_values = [values[i] for i in sorted_indices]
```

```python
#plotting
plt.figure(figsize=(8, 6))
plt.bar(sorted_categories, sorted_values)

#labels
for i, v in enumerate(sorted_values):
    plt.text(i, v + 0.5, f"{v:.2f}%", ha='center', fontsize=12, fontweight='bold')

#labels and title
plt.ylabel("Percentage Scoring 80+")
plt.title("Advanced Regents Performance Comparison (80+)")
```

Conclusion:
When comparing the percentage of students who scored at an advanced level (80+) on their Regents exams, you can see the following:

John Adams had 13.22% of its students scoring at an advanced level, which is the lowest among the three categories.
Queens saw 29.38% of students scoring 80+ on the Regents exams, coming in highest when compared to John Adams as well as the Five Boroughs.
The Five Boroughs averaged at 22.69% of students scoring 80+.
To sum it all up, even though students from Queens showed above average in advanced performance exam scores when compared to the Five Boroughs, students from John Adams with advanced scores were significantly less on average overall.
